```python
import numpy as np
import cv2
import os
```

```python
images_file_path='./dataset/Hands'
output_file_path='./dataset/Images_Afterpreprocess'
images_name=os.listdir(images_file_path)
```

```python
if not os.path.exists(output_file_path):
    os.makedirs(output_file_path)
```

```python
def preprocessing(I):
    # Load the original RGB image
    image = I

    # Convert the image to floating-point representation
    image_float = image.astype(np.float32) / 255.0
  
    # Split the image into individual channels (B, G, R)
    b, g, r = cv2.split(image_float)

    # Define the guided filter parameters
    radius = 10
    eps = 0.01

    # Apply the guided filter to each channel
    filtered_r = cv2.ximgproc.guidedFilter(r, r, radius, eps)
    filtered_g = cv2.ximgproc.guidedFilter(g, g, radius, eps)
    filtered_b = cv2.ximgproc.guidedFilter(b, b, radius, eps)

    # Merge the filtered channels back into an RGB image
    filtered_image = cv2.merge([filtered_b, filtered_g, filtered_r])
    
    #resize image
    result = cv2.resize(filtered_image, (224,224))
    
    # Generate gray image
    result = cv2.cvtColor(result, cv2.COLOR_BGR2GRAY)
    result = np.uint8(result*255.0)
    return result
```

```python
count=1
for i in images_name:
    I = cv2.imread(images_file_path+'/'+i)
    result = preprocessing(I)
    cv2.imwrite(output_file_path+'/'+i, result)
    if count%100==0:
        print(str(count)+' images are pre-processed')
        print('======================================')
    count+=1
```
