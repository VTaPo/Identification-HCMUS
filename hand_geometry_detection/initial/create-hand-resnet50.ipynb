```python
import torch
from torchvision.models import resnet

model = resnet.resnet50(pretrained=True)  # Instantiate the ResNet model
```

```python
num_classes = 189  # Replace with your desired number of classes

# Modify the last layer of the model
in_features = model.fc.in_features
model.fc = torch.nn.Linear(in_features, num_classes)
```

```python
torch.save(model.state_dict(), './hand_resnet101.pth')
```
